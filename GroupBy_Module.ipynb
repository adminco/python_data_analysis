### The .agg() method

This method allows us to combine different aggregate operations all in one table

```python
sectors.agg(["sum", "mean", "min"])
```

If you just want specific aggregrate operations on different fields, the agg is passed a dictionary (key-value paris)

```python
sectors.agg({"Revenue" : "mean",
            "Profits": "max",
            "Employees": "sum"})
#the above gets the mean for Revenue field; max for Profits tab and so on
```

### Looping through Groups

For example, lets say we want to get all the companies with maximum revenues in each group and store them in a DataFrame:

```python
df = pd.DataFrame(columns= fort.columns)
#build the dataFrame you want to store the data in
```

```python
for sector, data in sectors:
    max_company = data.nlargest(1, "Revenue")
    df = df.append(max_company)

#this loops through the data sets in the sector group
#gets the max value after sorting by "Revenue" and storing the record in a variable
#append the data to the dataFrame built
```
